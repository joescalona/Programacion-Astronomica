Se definen dos funciones
$$ f(x) = x^2 - 2 $$
y su derivada, 
$$ f'(x) = 2x $$ 

```python
def f(x):
    return x**2 - 2

def df(x):
    return 2*x 
```

Importamos dos librerias útiles para graficar: **Numpy** y **Matplotlib**. Con la ayuda de Numpy podemos definir un vector que contengan los valores discretos que se graficarán luego con las funciones de matplotlib: 

```python
import numpy as np 
import matplotlib.pyplot as plt 

#comienza en -2 y termina en 2, con 98 numeros entre medio
x = np.linspace(-2,2,100)
plt.figure(num='Newton')
plt.plot(x,f(x))
plt.grid('on')
plt.show()
```

Aplicando ahora el **Método de Newton-Raphson** para encontrar una raíz de $f(x)$. Definimos el valor inicial = 4 y aplicaremos la derivada para una mejor aproximación. Estableceremos una diferencia mínima entre aproximaciones consecutivas para finalizar el código. 

```python
xNew = [4.0] #valor inicial
erro = 0.001
resp = xNew[-1]

print '%11s %11s' % ('x','f(x)')
while True:
    print '%11.8f % 11.8f' % (resp, f(resp))
    dres = -1 * f(resp)/df(resp)
    resp = resp + dres
    xNew.append(resp)
    if abs(xNew[-2]-xNew[-1] < erro):
        break
```

```python
xNew = np.array(xNew)
x = np.linspace(0,5,100)
plt.figure(num='Newton')
plt.clf()
plt.plot(x,f(x))
plt.plot(xNew, f(xNew),'ko',ms=6)
plt.grid('on')
plt.show()
```
