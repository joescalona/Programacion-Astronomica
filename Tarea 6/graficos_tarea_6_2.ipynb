```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
%matplotlib inline 
```

# Line plots 

## The basic 

```python
x = np.arange(-4*np.pi,4*np.pi,0.25)
sincx = np.sin(x)/x 
plt.plot(x,sincx)
plt.savefig('13.jpg')
```

```python
plt.plot(x,sincx,'ro')
plt.savefig('14.jpg')
```

```python
plt.plot(x,sincx,'k--')
plt.savefig('15.jpg')
```

```python
plt.plot(x,sincx,lw=10,alpha =0.5)
plt.plot(x,-sincx,'g--',lw = 5)
plt.savefig('16.jpg')
```

## Setting the axis range 

```python
plt.plot(x,sincx,lw=5,alpha =0.5)
plt.plot(x,-sincx,'g--',lw = 2.5)
plt.xlim([-4*np.pi,4*np.pi])
plt.ylim([-1.25,1.25])
plt.savefig('17.jpg')
```

## Adding labels and titles  

```python
plt.plot(x,sincx,lw=5,alpha =0.5)
plt.plot(x,-sincx,'g--',lw = 2.5)
plt.xlim([-4*np.pi,4*np.pi])
plt.ylim([-1.25,1.25])
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.title('The sinc function')
plt.savefig('18.jpg')
```

## Annotation and legends

```python
plt.plot(x,sincx,lw=5,alpha =0.5)
plt.plot(x,-sincx,'g--',lw = 2.5)
plt.xlim([-4*np.pi,4*np.pi])
plt.ylim([-1.25,1.25])
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.title('The sinc function')
plt.text(0,1.05,'Maximum',horizontalalignment='center')
plt.savefig('19.jpg')
```

```python
plt.plot(x,sincx,lw=5,alpha =0.5)
plt.plot(x,-sincx,'g--',lw = 2.5)
plt.xlim([-4*np.pi,4*np.pi])
plt.ylim([-1.25,1.25])
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.title('The sinc function')
plt.text(0,1.05,'Maximum',horizontalalignment='center')
plt.annotate('first minimum',(x[68],sincx[68]),xytext=(6,-0.75),arrowprops={'shrink':0.05,'facecolor':'black'})
plt.savefig('20.jpg')
```

```python
plt.plot(x,sincx,lw=5,alpha =0.5)
plt.plot(x,-sincx,'g--',lw = 2.5)
plt.xlim([-4*np.pi,4*np.pi])
plt.ylim([-1.25,1.25])
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.title('The sinc function')
plt.text(0,1.05,'Maximum',horizontalalignment='center')
plt.annotate('first minimum',(x[68],sincx[68]),xytext=(6,-0.75),arrowprops={'shrink':0.05,'facecolor':'black'})
plt.legend(('sincx(x)','-sinc(x)'))
plt.savefig('21.jpg')
```

## Filled plots

```python
plt.fill_between(x,np.sin(x/2),np.cos(x/2),alpha=0.5)
plt.savefig('22.jpg')
```

# Multiple Figures, and Multi-Panel Plots

## Multiple axes using subplot 

```python
plt.subplot(3,2,1)
plt.plot(x,sincx)
plt.subplot(3,2,2)
plt.plot(x,-sincx,'ro')
plt.savefig('23.jpg')
```

## Manually managing axes

```python
plt.subplot(3,2,1)
plt.plot(x,sincx)
plt.subplot(3,2,2)
plt.plot(x,-sincx,'ro')
plt.subplot(2,1,2)
plt.plot(x,sincx**2)
plt.savefig('24.jpg')
```
