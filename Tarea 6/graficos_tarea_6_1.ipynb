```python
x = np.linspace(0.,10.,50)
y = np.sin(x)
plt.plot(x,y)
plt.savefig('3.jpg')
```

Las funciones de `plot` son un poquito complejas, admite muchos parámetros, entre ellos: 

 - aplha: grado de transparencia, float(0.0 = transparente a 1.0 = opaco)
 - color o c: color de matplotlib 
 - label: Etiqueta con cadena de texto, string. **Para escribir en latex, se debe introducir una r antes de las ' '** 
 - markered-gecolor o mec: Color de borde del símbolo 
 - markered-gewidth o mew: Ancho del borde del símbolo, float
 - markerfa-cecolor o mfc: Color del símbolo
 - markersize o ms: Tamaño del símbolo, float
 - linestyle o ls: Tipo de línea, "-" "--" "-." ":" "None"
 - linewidth o lw: Ancho de línea, float 
 - marker: Tipo de símbolo 
 
Ejemplo: 

```python
plt.plot(x,y , marker = 'o' , c='green', linewidth= 2)
#con linewidth = 0 Las lineas desaparecen
plt.savefig('4.jpg')
```

```python
plt.plot(x, y, marker= 'o', 
         color='green', 
         linewidth=0)
plt.savefig('5.jpg')
```

```python
plt.plot(x, y, 'go')
plt.savefig('6.jpg')
```
