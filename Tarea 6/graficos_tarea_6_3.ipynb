```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
%matplotlib inline
```

# Plotting 2D Data 

```python
x = np.arange(-2*np.pi,2*np.pi,0.01)
y = np.arange(-2*np.pi,2*np.pi,0.01)
xx,yy = np.meshgrid(x,y,indexing='ij')

r   = np.sqrt(xx**2 + yy**2)
phi = np.arccos(x/r)
z1  = np.sin(2*r)/r
z2  = z1 * (1+0.5*np.cos(4*phi))


```

## Contour plots


```python
plt.contour(xx,yy,z1)
plt.savefig('26.jpg')
```

```python
plt.contour(xx,yy,z1)
plt.gca().set_aspect('equal')
plt.draw()
plt.savefig('27.jpg')
```

```python
plt.cla()
plt.gca().set_aspect('equal')
plt.draw()
plt.contour(xx, yy, z1, np.arange(-0.4, 2.0, 0.1))
plt.savefig('28.jpg')
```

```python
plt.cla()
plt.gca().set_aspect('equal')
plt.draw()
cs = plt.contour(xx,yy,z1,np.arange(-0.4,2.0,0.4))
plt.clabel(cs)
plt.savefig('29.jpg')
```

```python
plt.cla()
plt.gca().set_aspect('equal')
plt.draw()
plt.contourf(xx,yy,z2,np.arange(-0.7,3.21,0.5))
plt.savefig('30.jpg')
```

```python
cs = plt.contour(xx,yy,z2,np.arange(-0.7,3.21,0.5),colors='k')
plt.gca().set_aspect('equal')
plt.draw()
plt.clabel(cs)
plt.savefig('31.jpg')
```

```python
plt.clf()
plt.imshow(z2,aspect='equal',origin='lower',extent=(-2*np.pi,2*np.pi,-2*np.pi,2*np.pi))
plt.savefig('32.jpg')
```

```python
plt.clf()
plt.imshow(z2,aspect='equal',origin='lower',extent=(-2*np.pi,2*np.pi,-2*np.pi,2*np.pi))
plt.colorbar()
plt.savefig('33.jpg')
```

```python
plt.clf()
plt.imshow(z2,aspect='equal',origin='lower',extent=(-2*np.pi,2*np.pi,-2*np.pi,2*np.pi),vmin=-1,vmax=3)
plt.colorbar()
plt.savefig('34.jpg')
```

## Colors

```python
import matplotlib.cm as cm 
```

```python
plt.imshow(z2,aspect='equal',origin='lower',extent=(-2*np.pi,2*np.pi,-2*np.pi,2*np.pi),vmin=-1, vmax=3, cmap=cm.hot)
plt.colorbar()
plt.savefig('35.jpg')
```

# Plotting 3D Data 

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax=plt.gcf().add_subplot(111, projection='3d')
ax.plot_surface(xx,yy,z2)
plt.draw()
plt.savefig('36.jpg')
```

```python
ax=plt.gcf().add_subplot(111, projection='3d')
ax.plot_surface(xx,yy,z2,linewidth=0,cstride=50, rstride=50, cmap=cm.hot)
plt.savefig('37.jpg')
```

```python
plt.clf()
ax=plt.gcf().add_subplot(111, projection='3d')
ax.plot_wireframe(xx,yy,z2,lw=0.8,cstride=20, rstride=20)
plt.savefig('38.jpg')
```
